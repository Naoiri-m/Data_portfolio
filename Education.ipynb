### Pivot de dfp

```python
dfc= dfp.copy()
del dfc['Taux de mentions attendu - Toutes series']
print(f'Dimensin de dfc :{dfc.shape}.')
print(dfc.info())
```

L'objectif est de réduire le nmobre de colonnes en pivotant toute les colonnes de type 'Presents - {filière}', 'Taux de réussite - {filière}', 'Taux de mentions - {filière}' par quatre colonnes : 'Filière' :[L, ES,S, STG...], 'Présents' : [Nombre de présents dans la filière pour chaque établissement] 'Taux de réussite', Taux de mentions

```python
# Utiliser pd.melt pour transformer les colonnes désirées en lignes
dfc_long = pd.melt(dfc, 
                  id_vars=['Etablissement', 'Annee', 'Ville', 'UAI', 'Code commune','Academie','Departement', 'Secteur', 'Presents - Toutes series',
                          'Taux de reussite - Toutes series','Taux de reussite attendu acad - Toutes series','Taux de reussite attendu france - Toutes series',
                          'Taux de mentions - Toutes series', 'Effectif de seconde','Effectif de premiere', 'Effectif de terminale', 'Region',
                           'Code region', 'Code departement', 'Nombre de mentions TB avec félicitations - G', 'Nombre de mentions TB sans félicitations - G',
                           'Nombre de mentions B - G', 'Nombre de mentions AB - G', 'Nombre de mentions TB avec félicitations - T', 
                           'Nombre de mentions TB sans félicitations - T', 'Nombre de mentions B - T', 'Nombre de mentions AB - T'],
                   
                  value_vars=[col for col in dfc.columns if 'Presents' in col or 'Taux de reussite' in col or 'Taux de mentions' in col],
                   
                  var_name='variable',
                  value_name='value')
dfc_long
```

```python
# Séparer les colonnes en trois nouvelles colonnes: Filière, Type de donnée (Presents, Taux de reussite, Taux de mentions)
dfc_long['Filière'] = dfc_long['variable'].apply(lambda x: x.split(' - ')[1])
dfc_long['Type de donnée'] = dfc_long['variable'].apply(lambda x: x.split(' - ')[0])
dfc_long
#dfc_long['Type de donnée'].unique()
```

```python
print(dfc_long.columns)
fu = dfc_long['Filière'].unique()
tpu = dfc_long['Type de donnée'].unique()
print(f'Valeurs uniques de la colonne filière :{fu}.')
print(f'Valeurs uniques de la colonne Type de donnée: {tpu}.')
```

```python
# Pivot pour avoir les colonnes souhaitées: Présents, Taux de reussite, Taux de mentions
dfc_pivot = dfc_long.pivot_table(index=['Etablissement', 'Annee', 'Ville', 'UAI', 'Code commune','Academie','Departement', 'Secteur', 
                                        'Presents - Toutes series', 'Taux de reussite - Toutes series','Taux de reussite attendu acad - Toutes series',
                                        'Taux de reussite attendu france - Toutes series', 'Taux de mentions - Toutes series','Filière'],
                                 columns='Type de donnée',
                                 values='value',                                 
            ).reset_index()
print("\nDataFrame pivoté:\n", dfc_pivot)
dfc_pivot
```

**Problème : Dataframe vide**
Finalement le problème ne vient pas des valeurs manquantes. J'ai décidé d'ajouter au fur et à mesure les colonnes dans 'index', afin d'identifier la/les colonne(s) qui crée(nt) la dataframe vide. Pour chaque ajout la dataframe sortait correctement, avec les colonnes 'Filière', 'Taux de réussite' et 'Taux de mentions' jusqu'au moment de l'ajout des colonnes `Taux de mentions - Toutes series`, `Effectif de terminale`, `Nombre de mentions TB avec félicitations - G`, `Nombre de mentions TB sans félicitations - G`, `Nombre de mentions B - G`, `Nombre de mentions AB- G`, `Nombre de mentions TB avec félicitations - T`, `Nombre de mentions TB sans félicitations - T`, `Nombre de mentions B - T`, `Nombre de mentions AB - T`.

Raisons possibles :
Valeurs manquantes ou incohérentes : Si la colonne 'Taux de mentions - Toutes series' contient des valeurs manquantes ou des valeurs non cohérentes pour certaines combinaisons de l'index, cela peut empêcher la création de lignes correspondantes dans le pivot.

Doublons ou incompatibilités dans les données : Il se peut qu'il y ait des doublons ou des valeurs qui ne correspondent pas aux autres colonnes d'index, ce qui entraîne un conflit lors du pivot.

```python
# Vérification des valeurs uniques dans 'Taux de mentions - Toutes series'
print("Valeurs uniques dans 'Taux de mentions - Toutes series':\n", dfc_long['Taux de mentions - Toutes series'].unique())

# Vérification du nombre de valeurs manquantes dans cette colonne
print("\nValeurs manquantes dans 'Taux de mentions - Toutes series':", dfc_long['Taux de mentions - Toutes series'].isna().sum())
```

```python
# Compter les combinaisons de valeurs uniques dans les colonnes d'index avant le pivot
comb_before = dfc_long.groupby(['Etablissement', 'Annee', 'Ville', 'UAI', 'Code commune', 'Academie', 'Departement', 
                                 'Secteur', 'Presents - Toutes series', 'Taux de reussite - Toutes series', 
                                 'Taux de reussite attendu acad - Toutes series', 'Taux de reussite attendu france - Toutes series', 
                                 'Filière']).size()

print("\nNombre de combinaisons uniques avant l'ajout de 'Taux de mentions - Toutes series':", comb_before.shape)

# Compter les combinaisons après avoir ajouté 'Taux de mentions - Toutes series'
comb_after = dfc_long.groupby(['Etablissement', 'Annee', 'Ville', 'UAI', 'Code commune', 'Academie', 'Departement', 
                                'Secteur', 'Presents - Toutes series', 'Taux de reussite - Toutes series', 
                                'Taux de reussite attendu acad - Toutes series', 'Taux de reussite attendu france - Toutes series', 
                                'Taux de mentions - Toutes series', 'Filière']).size()

print("\nNombre de combinaisons uniques après l'ajout de 'Taux de mentions - Toutes series':", comb_after.shape)
```

Le fait que le nombre de combinaisons uniques tombe à zéro après l'ajout de la colonne 'Taux de mentions - Toutes series' indique clairement qu'il y a un problème lié à cette colonne. Pour trouver la source du problème nous suivrons les étapes suivantes :

```python
# Voir la distribution des valeurs manquantes dans 'Taux de mentions - Toutes series'
missing_mentions = dfc_long['Taux de mentions - Toutes series'].isna().groupby(
    [dfc_long['Etablissement'], dfc_long['Annee'], dfc_long['Ville'], dfc_long['Filière']]
).sum()

print("\nDistribution des valeurs manquantes par établissement, année, ville et filière:\n", missing_mentions)
```

```python
# Vérifier les lignes où 'Taux de mentions - Toutes series' n'est pas manquant
valid_mentions = dfc_long.dropna(subset=['Taux de mentions - Toutes series'])

# Compter les combinaisons uniques dans ces lignes valides
comb_valid = valid_mentions.groupby(['Etablissement', 'Annee', 'Ville', 'UAI', 'Code commune', 'Academie', 
                                     'Departement', 'Secteur', 'Presents - Toutes series', 
                                     'Taux de reussite - Toutes series', 'Taux de reussite attendu acad - Toutes series', 
                                     'Taux de reussite attendu france - Toutes series', 'Filière']).size()

print("\nNombre de combinaisons uniques valides (avec valeurs non manquantes):", comb_valid.shape)
```

Le fait que le nombre de combinaisons uniques valides soit égal à zéro signifie qu'il n'existe aucune ligne dans votre DataFrame où toutes les colonnes d'index et la colonne 'Taux de mentions - Toutes series' aient des valeurs non manquantes simultanément.

**Remplacement des valeurs manquantes**

`Taux de mentions - Toutes series` = (somme du nombres de mentions) / effectif  de terminale

`Effectif de terminale` = Presents - Toutes series

Concernant le `nombre de mentions`, c'est un chiffre qu'on ne peux calculer de manière précise et correcte. Nous allons donc partir du principe que lorque qu'il n'y a pas de valeurs c'est qu'il n'y a pas eu de mentions comptabilisées.
Pour remplir ces champs manquants je vais retravailler sur la dataframe dfc.

```python
#Commençon par "Effectif de terminale"

# Afficher les valeurs manquantes avant remplacement
missing_edt = dfc['Effectif de terminale'].isna().sum()
print(f"Valeurs manquantes dans 'Effectif de terminale' avant remplacement: {missing_edt}")

# Remplacer les valeurs manquantes dans 'Effectif de terminale' par les valeurs de 'Presents - Toutes series'
dfc['Effectif de terminale'] = dfc['Effectif de terminale'].fillna(dfc['Presents - Toutes series'])

# Afficher les valeurs manquantes après remplacement
missing_edt = dfc['Effectif de terminale'].isna().sum()
print(f"Valeurs manquantes dans 'Effectif de terminale' après remplacement: {missing_edt}")

# Vérifier les premières lignes pour s'assurer que le remplacement a fonctionné
dfc[['Effectif de terminale', 'Presents - Toutes series']]
```

```python
# Liste des colonnes à traiter
colonnes_mentions = [
    'Nombre de mentions TB avec félicitations - G',
    'Nombre de mentions TB sans félicitations - G',
    'Nombre de mentions B - G',
    'Nombre de mentions AB - G',
    'Nombre de mentions TB avec félicitations - T',
    'Nombre de mentions TB sans félicitations - T',
    'Nombre de mentions B - T',
    'Nombre de mentions AB - T'
]

# Compter les valeurs manquantes avant remplacement
nb_valeurs_manquantes_avant = dfc[colonnes_mentions].isna().sum().sum()
print(f"Nombre total de valeurs manquantes avant remplacement: {nb_valeurs_manquantes_avant}")

# Remplacer les valeurs manquantes par 0 dans les colonnes mentionnées
dfc[colonnes_mentions] = dfc[colonnes_mentions].fillna(0)

# Compter les valeurs manquantes après remplacement
nb_valeurs_manquantes_apres = dfc[colonnes_mentions].isna().sum().sum()
print(f"Nombre total de valeurs manquantes après remplacement: {nb_valeurs_manquantes_apres}")
```

```python
### Valeurs manquantes 'Taux de mentions - Toutes series'

missing_tmts = dfc['Taux de mentions - Toutes series'].isna().sum()
print(f"Valeurs manquantes dans 'Taux de mentions - Toutes series' avant calcul: {missing_tmts}")

# Calculer la somme des mentions pour chaque ligne
dfc['Somme_mentions'] = (
    dfc['Nombre de mentions TB avec félicitations - G'] +
    dfc['Nombre de mentions TB sans félicitations - G'] +
    dfc['Nombre de mentions B - G'] +
    dfc['Nombre de mentions AB - G'] +
    dfc['Nombre de mentions TB avec félicitations - T'] +
    dfc['Nombre de mentions TB sans félicitations - T'] +
    dfc['Nombre de mentions B - T'] +
    dfc['Nombre de mentions AB - T']
)

# Calculer le taux de mentions là où il manque en divisant la somme par le nombre de présents
dfc['Taux de mentions - Toutes series calculé'] = (dfc['Somme_mentions'] / dfc['Presents - Toutes series']*100).round()

# Remplacer les valeurs manquantes dans la colonne 'Taux de mentions - Toutes series' par les valeurs calculées
dfc['Taux de mentions - Toutes series'] = dfc['Taux de mentions - Toutes series'].fillna(dfc['Taux de mentions - Toutes series calculé'])

# Compter les valeurs manquantes après le calcul
missing_tmts = dfc['Taux de mentions - Toutes series'].isna().sum()
print(f"Valeurs manquantes dans 'Taux de mentions - Toutes series' après calcul: {missing_tmts}")

# Supprimer la colonne intermédiaire 'Taux de mentions - Toutes series calculé' et 'Somme
dfc.drop(columns=['Taux de mentions - Toutes series calculé', 'Somme_mentions'], inplace=True)

# Vérifier les premières lignes pour s'assurer que le remplacement a fonctionné
#dfc[['Taux de mentions - Toutes series', 'Presents - Toutes series']].head(10)
dfc[dfc['Taux de mentions - Toutes series']!=0]
```

**Solution** : 

Afin de lever la limite du nombre de résultats lors de la récupération des données à partir de l'API, il faut utiliser la pagination fournie par l'API.
Pour cela, je vais utiliser la pagination via les paramètres `offset` et `limit` pour paginer à travers les résultats. Le code suivant effectue des requêtes successives à l'API, en augmentant l'offset à chaque fois pour récupérer des ensembles successifs de `100` résultats jusqu'à ce que tous les résultats soient récupérés. 

```python
# Nombre total d'enregistrements à récupérer
total_records = 312

# Nombre d'enregistrements par page
limit = 100

# Calcul du nombre total de pages nécessaires
total_pages = (total_records + limit - 1) // limit

# Liste pour stocker toutes les données récupérées
data_es = []

# Parcourir toutes les pages
for index in range(total_pages):
    offset = index * limit
    url = f'https://data.education.gouv.fr/api/explore/v2.1/catalog/datasets/fr-en-reussite-au-baccalaureat-origine-sociale/records?limit={limit}&offset={offset}'
    
    # Faire une requête GET à l'API
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        
        # Vérifier si la clé 'results' est présente dans les données
        if 'results' in data:
            results = data['results']
            
            # Extraire les champs des enregistrements et les ajouter à la liste
            data_es.extend(record for record in results)
        else:
            print(f"La clé 'results' n'a pas été trouvée dans les données pour l'offset {index}.")
    else:
        print(f"Échec de la récupération des données pour l'offset {index}, code de statut : {response.status_code}")
        break  # Sortir de la boucle en cas d'erreur pour éviter de faire des demandes supplémentaires

# Une fois que tous les enregistrements sont récupérés, convertir la liste de données en DataFrame
dos = pd.DataFrame(data_es)

# Afficher les premières lignes du DataFrame
print("Premières lignes du DataFrame : ")
print(dos.head())

# Afficher le nombre total d'enregistrements récupérés
print(f"Nombre total d'enregistrements récupérés : {len(dos)}")
```

### Webscraping : "Le classement des revenus par ville" 

```python
import requests
from bs4 import BeautifulSoup
```

```python
def scrape_year(year):
    # Initialiser une liste pour stocker les données de l'année
    year_data = []

    page = 1
    while True:
        # Construire l'URL en fonction de l'année et de la page
        url = f'https://www.journaldunet.com/business/salaire/classement/villes/revenus/{year}?page={page}'
        
        # Envoyer une requête HTTP pour obtenir le contenu de la page
        response = requests.get(url)
        
        # Si le statut est 404, arrêter la boucle
        if response.status_code == 404:
            print(f"Fin des pages pour l'année {year} à la page {page-1}")
            break
        
        # Si le statut n'est pas 200, signaler l'erreur et arrêter
        if response.status_code != 200:
            print(f"Impossible de récupérer la page {page} pour l'année {year}, statut HTTP:", response.status_code)
            break
        
        # Parser le contenu HTML
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Trouver le tableau avec la classe "odTable"
        table = soup.find('table', {'class': 'odTable'})
        
        # Vérifier si le tableau existe sur la page
        if not table:
            print(f"Pas de tableau trouvé pour l'année {year}, page {page}")
            break
        
        # Récupérer les lignes du tableau
        rows = table.find('tbody').find_all('tr')
        for row in rows:
            cells = row.find_all('td')
            # Extraire les données et ajouter l'année
            row_data = [year] + [cell.text.strip().replace('\xa0', ' ') for cell in cells]
            year_data.append(row_data)
        
        # Passer à la page suivante
        page += 1
    
    # Créer un DataFrame pour l'année
    df_year = pd.DataFrame(year_data, columns=['Année', 'Rang', 'Ville', 'Revenu fiscal moyen'])
    
    return df_year
```
